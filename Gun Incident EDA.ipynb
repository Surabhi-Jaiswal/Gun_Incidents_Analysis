```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Reason', data=df)
plt.title('Incident Reasons Distribution')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Age'].dropna(), bins=30, kde=True)
plt.title('Age Distribution')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Race', data=df)
plt.title('Race Distribution')
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(x='Sex', data=df)
plt.title('Sex Distribution')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Reason', hue='Sex', data=df)
plt.title('Reason vs Sex')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Reason', y='Age', data=df)
plt.title('Age Distribution across Reasons')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Place of incident', hue='Reason', data=df)
plt.title('Place of Incident vs Reason')
plt.xticks(rotation=45)
plt.show()
```

```python
# Police involvement vs. Reason
plt.figure(figsize=(10, 6))
sns.countplot(x='Police involvement', hue='Reason', data=df)
plt.title('Police Involvement vs Reason')
plt.show()
```

```python
pivot_table = df.pivot_table(index='Month', columns='Year', aggfunc='size')

plt.figure(figsize=(10, 6))
sns.heatmap(pivot_table, cmap='Blues', annot=True, fmt="d")
plt.title('Incidents by Year and Month')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.violinplot(x='Reason', y='Age', data=df, palette='muted')
plt.title('Age Distribution by Reason')
plt.show()
```

```python
race_reason = df.groupby(['Race', 'Reason']).size().unstack()

race_reason.plot(kind='bar', stacked=True, figsize=(10, 6), colormap='viridis')
plt.title('Incident Reason by Race')
plt.ylabel('Count of Incidents')
plt.xticks(rotation=45)
plt.show()
```

```python
place_counts = df['Place of incident'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(place_counts, labels=place_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('Set3'))
plt.title('Distribution of Incidents by Place')
plt.show()
```
